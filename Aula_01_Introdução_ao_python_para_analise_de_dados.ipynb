Operadores s√£o s√≠mbolos que indicam opera√ß√µes a serem realizadas sobre valores ou vari√°veis. Em Python, os principais operadores num√©ricos s√£o:

## 8. Fun√ß√µes em Python e passagem de par√¢metros

Fun√ß√µes s√£o blocos de c√≥digo reutiliz√°veis que realizam uma tarefa espec√≠fica.
Elas permitem organizar melhor o programa, evitar repeti√ß√£o de c√≥digo e estruturar an√°lises cient√≠ficas de forma modular.

A sintaxe b√°sica √©:

```python

    return resultado
```

Elementos importantes:

* `def` inicia a defini√ß√£o.
* O nome deve ser descritivo.
* Par√¢metros s√£o valores recebidos pela fun√ß√£o.
* `return` devolve um valor (opcional, mas muito √∫til).
* O bloco interno deve ser indentado.

### Explica√ß√£o

* A fun√ß√£o recebe a vari√°vel `temp_c`.
* Realiza o c√°lculo dentro do corpo da fun√ß√£o.
* Retorna o valor convertido, que pode ser usado em qualquer parte do c√≥digo.

Fun√ß√µes assim s√£o √∫teis para converter unidades, normalizar medidas, limpar dados ou encapsular um c√°lculo repetitivo.

### 8.2 Fun√ß√£o com m√∫ltiplos par√¢metros

Fun√ß√µes podem receber quantos par√¢metros forem necess√°rios.

Exemplo: c√°lculo de tens√£o mec√¢nica.


### Explica√ß√£o

Esta fun√ß√£o organiza um c√°lculo muito comum em materiais:

* Recebe for√ßa e √°rea.
* Faz convers√£o interna.
* Devolve a tens√£o em Pascal.

Separar esse c√°lculo em uma fun√ß√£o torna o c√≥digo mais leg√≠vel e f√°cil de reutilizar.

### 8.3 Fun√ß√µes que n√£o retornam valor

Nem sempre √© necess√°rio usar `return`.
Fun√ß√µes podem apenas executar algo, como imprimir resultados.


### 8.5 Vari√°veis locais vs. vari√°veis globais

Em Python:

* **vari√°veis locais** existem somente dentro da fun√ß√£o.

   * Uma lista √© uma cole√ß√£o de objetos gen√©ricos.
   * Cada elemento ocupa espa√ßo extra e est√° espalhado pela mem√≥ria.
   * Opera√ß√µes como soma, multiplica√ß√£o ou vetor-matriz s√£o feitas elemento por elemento, via loops em Python, o que √© lento.

2. **C√°lculo cient√≠fico exige vetores e matrizes verdadeiras**, como em C/Fortran/Matlab.
   NumPy fornece isso atrav√©s do objeto central **ndarray**.

3. **NumPy usa c√≥digo compilado em C e Fortran**, muito mais r√°pido que loops em Python puro.
   Isso o torna ideal para:

   * an√°lise de dados experimentais
   * simula√ß√µes num√©ricas
   * processamento de matrizes de imagens de microscopia (SEM, TEM)

   * dados espectrosc√≥picos
   * difra√ß√£o de raio-X

4. **NumPy padroniza o ecossistema cient√≠fico de Python**, permitindo interc√¢mbio entre bibliotecas.

### 9.2 O que o NumPy oferece (vis√£o geral)

Sem listar tudo, estes s√£o pilares essenciais que justificam sua exist√™ncia:

* **Array multidimensional (ndarray)**
  Estrutura cont√≠nua na mem√≥ria, equivalente a vetores/matrizes de C e Fortran.

* **Opera√ß√µes vetorizadas**
  Opera√ß√µes como soma, multiplica√ß√£o, logaritmos, ra√≠zes e fun√ß√µes trigonom√©tricas s√£o aplicadas *a todo o array* sem loops expl√≠citos.

* **Broadcasting**
  Permite opera√ß√µes entre arrays de formas diferentes sem duplica√ß√£o de dados.

* **√Ålgebra linear (via `numpy.linalg`)**
  Determinantes, autovalores, sistemas lineares, decomposi√ß√µes.

* **Opera√ß√µes estat√≠sticas r√°pidas**
  M√©dias, desvios, somat√≥rios, correla√ß√µes.

* **Gera√ß√£o de n√∫meros aleat√≥rios**
  Importante em simula√ß√£o, experimentos Monte Carlo e tratamentos estat√≠sticos.

* **Manipula√ß√£o eficiente de formas e transposi√ß√µes**
  Essencial em tensores, campos discretizados, matrizes de dados e imagens cient√≠ficas.

* **Integra√ß√£o com bibliotecas cient√≠ficas**
  Quase tudo em ci√™ncia dos materiais usa NumPy internamente:

# 13. Gr√°fico final: Temperatura √ó Tempo em um ensaio t√©rmico

Agora que todos os elementos foram apresentados, constr√≥i-se o gr√°fico completo do experimento hipot√©tico de varia√ß√£o de temperatura ao longo do tempo.

## Explica√ß√£o do gr√°fico final

* O marcador torna vis√≠veis os pontos individuais do experimento.
* A grade auxilia na leitura espec√≠fica de valores.
* A figura cont√©m t√≠tulo, r√≥tulos e formata√ß√£o adequada para relat√≥rios t√©cnicos.

Este tipo de gr√°fico √© utilizado em:

* an√°lises de tratamentos t√©rmicos,
