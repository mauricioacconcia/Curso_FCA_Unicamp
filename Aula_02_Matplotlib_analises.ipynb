## 4. Primeira visualização científica

### 4.1 Estrutura básica de um gráfico
A construção de um gráfico segue etapas fundamentais: criação de uma figura, plotagem dos dados, configuração dos rótulos e exibição final. Cada uma dessas etapas compõe a estrutura de uma visualização clara e interpretável.

Um gráfico sem rótulos não comunica adequadamente as informações.
```python
# Importação das bibliotecas essenciais
import numpy as np              # NumPy: operações numéricas rápidas com vetores e matrizes
import pandas as pd             # Pandas: manipulação de dados tabulares (será usado mais à frente)
import matplotlib.pyplot as plt # Matplotlib: criação de gráficos científicos

# Configuração opcional do estilo visual dos gráficos
plt.style.use('seaborn-v0_8-darkgrid')

# Verificação das versões instaladas (útil para garantir compatibilidade no Colab)
print("NumPy versão:", np.__version__)
print("Pandas versão:", pd.__version__)

# Criação de um vetor numérico com 100 valores igualmente espaçados entre 0 e 10
x = np.linspace(0, 10, 100)

# Aplicação de uma função matemática simples (seno) sobre todos os pontos
y = np.sin(x)

# Exibição dos primeiros valores para inspeção
print("Primeiros valores de x:", x[:5])
print("Primeiros valores de y:", y[:5])

# Construção do primeiro gráfico
plt.figure(figsize=(8, 4))          # Define tamanho da área do gráfico
plt.plot(x, y, label='sin(x)')      # Desenha a curva seno
plt.xlabel('x')                     # Rótulo do eixo x
plt.ylabel('sin(x)')                # Rótulo do eixo y
plt.title('Primeiro gráfico em Python')  # Título do gráfico
plt.legend()                        # Exibe legenda configurada
plt.show()                          # Renderiza o gráfico no notebook
```

#NumPy aplicado a dados científicos

## 1. Introdução ao papel do NumPy em dados científicos
NumPy é a base de praticamente todo o ecossistema científico em Python. Ele fornece estruturas de dados otimizadas para operações numéricas, permitindo manipular vetores e matrizes com desempenho muito superior ao obtido com listas nativas do Python. ---

## 2. Estruturas fundamentais: arrays unidimensionais e bidimensionais

### 2.1 Arrays unidimensionais

- perfis de propriedades em uma única direção (como dureza ao longo de uma linha).

Os arrays são criados rapidamente e podem ser transformados com facilidade, permitindo cálculos matemáticos diretos sobre todos os elementos.

### 2.2 Arrays bidimensionais
Arrays bidimensionais podem ser entendidos como tabelas numéricas, similares a matrizes. Esse tipo de estrutura aparece em diversas situações da engenharia de materiais, como:
- mapas de microdureza;
- matrizes de intensidade de imagens microscópicas convertidas para escala numérica;
- curvas múltiplas empilhadas (por exemplo, várias amostras de tração);
- superfícies de resposta obtidas por simulação.

A capacidade de manipular arrays 2D rapidamente permite realizar transformações, filtragens e cálculos sobre dados experimentais de forma eficiente.

O uso de pontos igualmente espaçados é importante para gerar curvas suaves e adequadas para visualização gráfica.

NumPy possui funções matemáticas vetorizadas para:

## 4. Operações numéricas fundamentais

### 4.1 Estatísticas básicas
Operações como média, desvio padrão e correlação são essenciais para interpretar dados experimentais. NumPy implementa essas operações de forma rápida e precisa. Em engenharia de materiais, elas são frequentemente usadas para:

### 4.2 Transformações matemáticas

### 4.3 Combinação e reorganização de dados

---

## 5. Integração com visualização científica
Todo o processamento numérico feito com NumPy pode ser enviado diretamente para o Matplotlib, facilitando a criação de gráficos. Curvas, superfícies e matrizes numéricas são convertidas de forma automática em elementos gráficos. Essa integração permite transformar resultados numéricos em visualizações interpretáveis imediatamente.



```python
# -----------------------------------------------------------
# Bloco 2 — NumPy aplicado a dados científicos (Código)
# -----------------------------------------------------------

# Importação das bibliotecas essenciais
import numpy as np
import matplotlib.pyplot as plt

# Estilo visual dos gráficos
plt.style.use('seaborn-v0_8-darkgrid')

# -----------------------------------------------------------
# 1. Criação de arrays unidimensionais
# -----------------------------------------------------------

# Vetor de 100 pontos igualmente espaçados entre 0 e 10
x = np.linspace(0, 10, 100)

# Aplicação de funções matemáticas vetorizadas
y_sin = np.sin(x)              # Função seno
y_exp = np.exp(-0.3 * x)       # Expoencial decrescente
y_comb = y_sin * y_exp         # Combinação das duas funções

# -----------------------------------------------------------
# 2. Criação de arrays bidimensionais (simulando mapas 2D)
# -----------------------------------------------------------

# Criando uma grade 2D usando meshgrid
x2 = np.linspace(-3, 3, 200)
y2 = np.linspace(-3, 3, 200)
X, Y = np.meshgrid(x2, y2)

# Simulando um mapa gaussiano (útil para representar picos de DRX ou mapas de fases)
Z = np.exp(-(X**2 + Y**2))

# -----------------------------------------------------------
# 3. Estatísticas fundamentais usando NumPy
# -----------------------------------------------------------

media = np.mean(y_comb)        # Média da curva combinada
desvio = np.std(y_comb)        # Desvio padrão
ponto_max = np.max(y_comb)     # Valor máximo
indice_max = np.argmax(y_comb) # Índice do valor máximo

# -----------------------------------------------------------
# 4. Visualizações básicas
# -----------------------------------------------------------

# Gráfico das curvas simuladas
plt.figure(figsize=(10, 4))
plt.plot(x, y_sin, label='sin(x)')
plt.plot(x, y_exp, label='exp(-0.3x)')
plt.plot(x, y_comb, label='sin(x) * exp(-0.3x)')
plt.xlabel('x')
plt.ylabel('valor')
plt.title('Funções simuladas com NumPy')
plt.legend()
plt.show()

# Visualização do mapa 2D
plt.figure(figsize=(6, 5))
plt.imshow(Z, extent=[-3, 3, -3, 3], origin='lower', cmap='viridis')
plt.colorbar(label='Intensidade')
plt.title('Mapa Gaussiano 2D Simulado')
plt.xlabel('x')
plt.ylabel('y')
plt.show()

# -----------------------------------------------------------
# 5. Impressão dos resultados estatísticos
# -----------------------------------------------------------

print("Média da curva combinada:", media)
print("Desvio padrão:", desvio)
print("Valor máximo:", ponto_max)
print("Índice do valor máximo:", indice_max)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # necessário para 3D

plt.style.use('seaborn-v0_8-darkgrid')

# -----------------------------------------------------------
# 1. Dados simulados básicos para gráficos 2D
# -----------------------------------------------------------

# Domínio contínuo
x = np.linspace(0, 10, 200)

# Curvas científicas simuladas
y1 = np.sin(x)                         # curva periódica
y2 = np.exp(-0.3 * x)                  # decaimento exponencial
y3 = y1 * y2                           # seno amortecido

# -----------------------------------------------------------
# 2. Gráfico de linha (curvas 2D)
# -----------------------------------------------------------

plt.figure(figsize=(10, 4))
plt.plot(x, y1, label='sin(x)')
plt.plot(x, y2, label='exp(-0.3x)')
plt.plot(x, y3, label='sin(x) * exp(-0.3x)')
plt.xlabel('x')
plt.ylabel('Valor')
plt.title('Gráficos de linha — Comportamentos típicos')
plt.legend()
plt.show()

# -----------------------------------------------------------
# 3. Gráfico de dispersão (simulando tamanho de partícula)
# -----------------------------------------------------------

np.random.seed(0)
tamanho = np.random.normal(loc=50, scale=8, size=200)   # distribuição normal
dureza = 200 + 0.5 * tamanho + np.random.normal(scale=5, size=200)  # relação linear + ruído

plt.figure(figsize=(7, 5))
plt.scatter(tamanho, dureza, alpha=0.7)
plt.xlabel('Tamanho de partícula (µm)')
plt.ylabel('Dureza (HV)')
plt.title('Dispersão — Relação entre tamanho de partícula e dureza')
plt.show()

# -----------------------------------------------------------
# 4. Histograma (distribuição de tamanhos)
# -----------------------------------------------------------

plt.figure(figsize=(7, 4))
plt.hist(tamanho, bins=20, color='royalblue', edgecolor='black')
plt.xlabel('Tamanho de partícula (µm)')
plt.ylabel('Frequência')
plt.title('Histograma — Distribuição de tamanho de partícula')
plt.show()

# -----------------------------------------------------------
# 5. Heatmap (mapa 2D simulado — microestrutura, fases)
# -----------------------------------------------------------

# Criando uma malha 2D
x2 = np.linspace(-3, 3, 200)
y2 = np.linspace(-3, 3, 200)
X, Y = np.meshgrid(x2, y2)

# Simulando um mapa de intensidade gaussiana
Z = np.exp(-(X**2 + Y**2))

plt.figure(figsize=(6, 5))
plt.imshow(Z, cmap='viridis', extent=[-3, 3, -3, 3], origin='lower')
plt.colorbar(label='Intensidade')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Heatmap — Mapa gaussiano simulado')
plt.show()

# -----------------------------------------------------------
# 6. Gráfico 3D — superfície de resposta
# -----------------------------------------------------------

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

Z_surface = np.sin(X) * np.cos(Y)     # superfície simulada

ax.plot_surface(X, Y, Z_surface, cmap='viridis', edgecolor='none')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Gráfico 3D — Superfície de resposta simulada')
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use('seaborn-v0_8-darkgrid')

# -----------------------------------------------------------
# 1. Criando um conjunto de dados tabular simulado (como se viesse de laboratório)
# -----------------------------------------------------------

# Simulando dados de um ensaio de tração
np.random.seed(42)

deformacao = np.linspace(0, 0.25, 300)                   # deformação entre 0% e 25%
modulo_elastico = 210e3                                  # MPa (exemplo aço)
tensao_elastica = modulo_elastico * deformacao           # comportamento elástico ideal
ruido = np.random.normal(0, 500, size=300)               # ruído experimental
tensao_real = tensao_elastica + ruido                    # curva real simulada

# Criando DataFrame
df = pd.DataFrame({
    "deformacao": deformacao,
    "tensao": tensao_real
})

# -----------------------------------------------------------
# 2. Manipulação com Pandas
# -----------------------------------------------------------

media_tensao = df["tensao"].mean()                       # média
desvio_tensao = df["tensao"].std()                       # desvio padrão

# Cálculo de uma nova coluna: tensão normalizada
df["tensao_normalizada"] = (df["tensao"] - media_tensao) / desvio_tensao

# -----------------------------------------------------------
# 3. Conversão para NumPy e operações numéricas avançadas
# -----------------------------------------------------------

# Extração de arrays NumPy
x_np = df["deformacao"].values
y_np = df["tensao_normalizada"].values

# Suavização simples por média móvel (janela 10)
janela = 10
suavizado = np.convolve(y_np, np.ones(janela)/janela, mode='same')

# -----------------------------------------------------------
# 4. Visualização — Curvas normalizada e suavizada
# -----------------------------------------------------------

plt.figure(figsize=(10, 4))
plt.plot(df["deformacao"], df["tensao_normalizada"], label="Normalizada")
plt.plot(df["deformacao"], suavizado, label="Suavizada (média móvel)", linewidth=2)
plt.xlabel("Deformação")
plt.ylabel("Tensão (normalizada)")
plt.title("Curva de tração simulada — Normalização e suavização")
plt.legend()
plt.show()

# -----------------------------------------------------------
# 5. Dispersão relacionando deformação e tensão real
# -----------------------------------------------------------

plt.figure(figsize=(7, 5))
plt.scatter(df["deformacao"], df["tensao"], s=10, alpha=0.6)
plt.xlabel("Deformação")
plt.ylabel("Tensão (MPa)")
plt.title("Dispersão — Dados reais simulados de tração")
plt.show()

# -----------------------------------------------------------
# 6. Histograma — Distribuição da tensão experimental
# -----------------------------------------------------------

plt.figure(figsize=(7, 4))
plt.hist(df["tensao"], bins=25, color="steelblue", edgecolor="black")
plt.xlabel("Tensão (MPa)")
plt.ylabel("Frequência")
plt.title("Histograma — Distribuição de tensão experimental simulada")
plt.show()

# -----------------------------------------------------------
# 7. Heatmap de correlação entre variáveis
# -----------------------------------------------------------

corr = df.corr(numeric_only=True)

plt.figure(figsize=(6, 5))
plt.imshow(corr, cmap="magma", interpolation="nearest")
plt.colorbar(label="Correlação")
plt.xticks(range(len(corr.columns)), corr.columns, rotation=45)
plt.yticks(range(len(corr.columns)), corr.columns)
plt.title("Heatmap — Matriz de correlação")
plt.tight_layout()
plt.show()






```

```python
# ============================================================
# 0. IMPORTS
# ============================================================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from mpl_toolkits.mplot3d import Axes3D

plt.style.use("seaborn-v0_8-darkgrid")

# ============================================================
# 1. LEITURA DAS BASES
# ============================================================
# Ajuste os caminhos só se o nome do arquivo estiver diferente
df_carbon = pd.read_csv(
    "/content/Steel- Property and Composition (Carbon Steel).csv",
    encoding="latin1"
)

df_stainless = pd.read_csv(
    "/content/Steel- Property and Composition (Stainless Steel).csv",
    encoding="latin1",
    engine="python",
    on_bad_lines="skip"  # pula linhas quebradas (títulos de família etc.)
)

print("==== PRIMEIRAS LINHAS – CARBON STEEL ====")
display(df_carbon.head())
print("==== PRIMEIRAS LINHAS – STAINLESS STEEL ====")
display(df_stainless.head())

# ============================================================
# 2. CONVERSÃO PARA VALORES NUMÉRICOS
#    (qualquer coisa não numérica vira NaN)
# ============================================================
df_carbon_num = df_carbon.copy()
df_stainless_num = df_stainless.copy()

for df in [df_carbon_num, df_stainless_num]:
    for col in df.columns:
        df[col] = pd.to_numeric(df[col], errors="coerce")

num_carbon = df_carbon_num.select_dtypes(include="number")
num_stainless = df_stainless_num.select_dtypes(include="number")

# ============================================================
# 3. ESTATÍSTICAS DESCRITIVAS
# ============================================================
print("===== ESTATÍSTICAS – CARBON STEEL =====")
display(num_carbon.describe())

print("===== ESTATÍSTICAS – STAINLESS STEEL =====")
display(num_stainless.describe())

# ============================================================
# 4. HEATMAPS DE CORRELAÇÃO
# ============================================================
plt.figure(figsize=(12, 8))
sns.heatmap(num_carbon.corr(), cmap="coolwarm")
plt.title("Heatmap de correlação – Carbon Steel")
plt.show()

plt.figure(figsize=(12, 8))
sns.heatmap(num_stainless.corr(), cmap="viridis")
plt.title("Heatmap de correlação – Stainless Steel")
plt.show()

# ============================================================
# 5. GRÁFICOS MECÂNICOS – CARBON STEEL
# ============================================================
plt.figure(figsize=(7, 5))
plt.scatter(df_carbon_num["YS (MPa)"], df_carbon_num["UTS (MPa)"], alpha=0.7)
plt.xlabel("Yield Strength YS (MPa)")
plt.ylabel("Ultimate Tensile Strength UTS (MPa)")
plt.title("YS × UTS – Carbon Steel")
plt.show()

plt.figure(figsize=(7, 5))
plt.scatter(df_carbon_num["Elongation (%)"], df_carbon_num["UTS (MPa)"], alpha=0.7)
plt.xlabel("Alongamento (%)")
plt.ylabel("UTS (MPa)")
plt.title("Ductilidade × Resistência – Carbon Steel")
plt.show()

plt.figure(figsize=(7, 5))
plt.scatter(df_carbon_num["Hardness (HB)"], df_carbon_num["UTS (MPa)"], alpha=0.7)
plt.xlabel("Dureza Brinell (HB)")
plt.ylabel("UTS (MPa)")
plt.title("Dureza × UTS – Carbon Steel")
plt.show()

# ============================================================
# 6. REGRESSÃO LINEAR – PREDIÇÃO DE UTS A PARTIR DA COMPOSIÇÃO
# ============================================================
features_reg = ["C (Max)", "Mn (Max)", "Si (Max)"]
X = df_carbon_num[features_reg]
y = df_carbon_num["UTS (MPa)"]

mask = (~X.isna().any(axis=1)) & (~y.isna())
X_reg = X[mask]
y_reg = y[mask]

model = LinearRegression()
model.fit(X_reg, y_reg)
y_pred = model.predict(X_reg)

print("\nCoeficientes da regressão para UTS (MPa):")
for name, coef in zip(features_reg, model.coef_):
    print(f"{name}: {coef:.4f}")
print(f"Intercepto: {model.intercept_:.4f}")

plt.figure(figsize=(7, 5))
plt.scatter(y_reg, y_pred, alpha=0.7)
plt.plot([y_reg.min(), y_reg.max()],
         [y_reg.min(), y_reg.max()],
         "r--")
plt.xlabel("UTS real (MPa)")
plt.ylabel("UTS previsto (MPa)")
plt.title("Regressão linear – UTS real × previsto")
plt.show()

# ============================================================
# 7. PCA – RELAÇÃO ENTRE COMPOSIÇÃO E PROPRIEDADES (CARBON)
# ============================================================
features_pca = ["C (Max)", "Mn (Max)", "Si (Max)",
                "UTS (MPa)", "YS (MPa)", "Hardness (HB)"]

df_pca = df_carbon_num[features_pca].dropna()

scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_pca)

pca = PCA(n_components=2)
pcs = pca.fit_transform(X_scaled)

plt.figure(figsize=(7, 5))
plt.scatter(pcs[:, 0], pcs[:, 1], alpha=0.7)
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("PCA – Carbon Steel (composição + propriedades)")
plt.show()

print("\nContribuição aproximada das variáveis em PC1:")
for name, comp in zip(features_pca, pca.components_[0]):
    print(f"{name}: {comp:.4f}")

# ============================================================
# 8. K-MEANS – AGRUPAMENTO DE LIGAS POR COMPORTAMENTO
# ============================================================
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
clusters = kmeans.fit_predict(X_scaled)

plt.figure(figsize=(7, 5))
plt.scatter(pcs[:, 0], pcs[:, 1], c=clusters, cmap="Accent", s=60)
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("Clusterização (K-means) – Carbon Steel")
plt.show()

# ============================================================
# 9. GRÁFICO 3D – COMPOSIÇÃO × RESISTÊNCIA (C, Mn, UTS)
# ============================================================
fig = plt.figure(figsize=(9, 7))
ax = fig.add_subplot(111, projection="3d")

ax.scatter(df_carbon_num["C (Max)"],
           df_carbon_num["Mn (Max)"],
           df_carbon_num["UTS (MPa)"],
           c=df_carbon_num["UTS (MPa)"],
           cmap="plasma")

ax.set_xlabel("Carbono C (Max)")
ax.set_ylabel("Manganês Mn (Max)")
ax.set_zlabel("UTS (MPa)")
ax.set_title("Efeito de C e Mn na UTS – Carbon Steel")
plt.show()

# ============================================================
# 10. ALGUNS GRÁFICOS ÚTEIS – STAINLESS STEEL
# ============================================================
plt.figure(figsize=(7, 5))
plt.scatter(df_stainless_num["Cr (Max)"], df_stainless_num["Ni (Max)"], alpha=0.7)
plt.xlabel("Cr (Max)")
plt.ylabel("Ni (Max)")
plt.title("Cr × Ni – Stainless Steel")
plt.show()

if "UTS (MPa)" in df_stainless_num.columns:
    plt.figure(figsize=(7, 5))
    plt.scatter(df_stainless_num["Cr (Max)"], df_stainless_num["UTS (MPa)"], alpha=0.7)
    plt.xlabel("Cr (Max)")
    plt.ylabel("UTS (MPa)")
    plt.title("Cr × UTS – Stainless Steel")
    plt.show()

plt.figure(figsize=(7, 5))
plt.boxplot([df_carbon_num["C (Max)"].dropna(),
             df_stainless_num["C (Max)"].dropna()],
            tick_labels=["Carbon", "Stainless"])
plt.ylabel("Carbono C (Max)")
plt.title("Comparação do teor de carbono – Carbon vs Stainless")
plt.show()

print("\nANÁLISE COMPLETA EXECUTADA.")
```

# 5. Regressão linear para prever UTS — *por que este modelo não é adequado*

Embora o código execute uma regressão linear utilizando as variáveis químicas  
**C (Max)**, **Mn (Max)** e **Si (Max)** para prever a resistência à tração (**UTS (MPa)**),  
é importante destacar que **o dataset não possui dados completos o suficiente** para que esse modelo seja estatisticamente válido.

A filtragem das amostras exige que todas as variáveis estejam preenchidas simultaneamente.  
Como as tabelas originais apresentam grande quantidade de valores faltantes  
(em especial nas colunas de composição), o número de linhas utilizáveis para regressão se reduz a apenas alguns poucos registros.

Isso implica que:

- o modelo é treinado sobre **uma amostra extremamente pequena**, insuficiente para capturar relações reais;
- os coeficientes obtidos **não representam tendências metalúrgicas confiáveis**;
- o gráfico “UTS real × UTS previsto” mostra poucos pontos, tornando impossível avaliar qualidade preditiva;
- qualquer aparente “coerência” é apenas efeito do baixo número de amostras, e não uma modelagem significativa.

Portanto, apesar de a regressão rodar tecnicamente, **ela não deve ser interpretada como um modelo válido**.  
A conclusão correta é que **os dados disponíveis não oferecem quantidade e completude suficientes para uma modelagem preditiva baseada na composição química**.

Análises mais apropriadas para este conjunto de dados são:

- estatísticas descritivas,
- correlações parciais,
- PCA,
- clusterização,
- gráficos mecânicos (UTS × dureza, UTS × ductilidade),
