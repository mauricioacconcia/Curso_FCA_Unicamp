O uso de aprendizado de máquina em ciência dos materiais tem se consolidado como uma ferramenta complementar aos métodos experimentais e teóricos tradicionais. A crescente disponibilidade de dados provenientes de ensaios mecânicos, caracterizações microestruturais e simulações computacionais cria um cenário favorável para a aplicação de técnicas capazes de identificar padrões, realizar previsões e apoiar a tomada de decisão científica. Nesta aula, o foco recai sobre uma visão aplicada, priorizando a compreensão do papel dessas técnicas no contexto real da pesquisa em materiais.

## 1. Motivação e contexto da aplicação de aprendizado de máquina em ciência dos materiais

A ciência dos materiais lida com sistemas complexos nos quais propriedades macroscópicas resultam da interação entre composição química, processamento e microestrutura. Métodos clássicos de modelagem nem sempre conseguem capturar relações não lineares ou de alta dimensionalidade presentes nesses dados. O aprendizado de máquina surge como uma abordagem capaz de explorar grandes volumes de dados experimentais e computacionais, auxiliando na previsão de propriedades, na classificação de estados microestruturais e na descoberta de correlações relevantes. Seu uso não substitui o conhecimento físico, mas atua como ferramenta de apoio à análise, interpretação e aceleração do desenvolvimento de novos materiais.

## 2. Conceitos fundamentais de aprendizado de máquina

Aprendizado de máquina é um conjunto de métodos computacionais destinados a construir modelos capazes de aprender relações a partir de dados e, com isso, produzir previsões, classificações ou estruturas descritivas úteis para análise. Em ciência dos materiais, o interesse raramente está no algoritmo em si, mas na capacidade do modelo de generalizar conhecimento a partir de medições e simulações, mantendo coerência com as restrições físicas do problema e permitindo interpretações que orientem decisões experimentais.

O ponto de partida é a definição do problema. Em termos formais, um conjunto de dados é composto por exemplos, e cada exemplo é descrito por atributos, também chamados de variáveis de entrada, características ou features. Esses atributos podem representar composição química, parâmetros de processamento, medidas extraídas de ensaios, descritores microestruturais ou variáveis de simulação. Quando existe uma variável alvo associada, como resistência à tração, dureza, condutividade, temperatura de transformação ou vida em fadiga, tem-se um problema supervisionado. O modelo tenta aprender uma função que mapeia os atributos para o alvo. Quando não há alvo explícito e o objetivo é organizar ou resumir o conjunto de dados, tem-se um problema não supervisionado. Há ainda o aprendizado por reforço, no qual um agente aprende por interação com um ambiente com base em recompensas, menos frequente em materiais, mas aplicável em otimização sequencial e planejamento de experimentos.

No aprendizado supervisionado, distinguem-se tarefas de regressão e de classificação. Regressão ocorre quando o alvo é contínuo, como módulo de elasticidade, tensão de escoamento ou taxa de crescimento de trinca. Classificação ocorre quando o alvo é categórico, como tipo de fratura, presença de defeito, classe de fase cristalina ou condição processual. Em ambos os casos, o objetivo central é generalização: o modelo deve funcionar em dados novos, não vistos durante o treinamento, e não apenas memorizar o conjunto disponível. Em materiais, generalização é especialmente crítica porque bases de dados podem ser pequenas, heterogêneas e com ruído experimental significativo.

A noção de modelo envolve parâmetros internos ajustados durante o treinamento. O treinamento é o processo de otimização desses parâmetros para minimizar uma função de perda, que quantifica o erro do modelo em relação ao que se deseja prever ou classificar. Para regressão, perdas típicas penalizam desvios numéricos, e para classificação penalizam atribuições incorretas de classe. A perda é minimizada por um algoritmo de ajuste, que depende do tipo de modelo. O resultado do treinamento é um conjunto de parâmetros que produz boas saídas em dados de treino, mas o critério prático de qualidade é o desempenho em dados de validação e teste, que aproximam o comportamento esperado em situações reais.

Uma distinção importante é entre parâmetros do modelo e hiperparâmetros. Parâmetros são aprendidos diretamente a partir dos dados, como coeficientes em regressões ou pesos em redes neurais. Hiperparâmetros são escolhas de configuração do algoritmo, como profundidade de árvores, número de árvores em uma floresta, valor de regularização em modelos lineares, tipo de kernel em SVM ou taxa de aprendizado em redes neurais. Em ciência dos materiais, a seleção de hiperparâmetros deve considerar não apenas desempenho numérico, mas também estabilidade, interpretabilidade e consistência com o domínio físico.

A avaliação do modelo é inseparável do desenho experimental do próprio ML. Separar dados em treino, validação e teste é essencial para estimar generalização. Em bases pequenas, usa-se frequentemente validação cruzada, na qual múltiplas partições são avaliadas e os resultados são agregados. Deve-se evitar vazamento de informação, que ocorre quando dados de teste influenciam treinamento ou seleção de hiperparâmetros, pois isso gera estimativas artificialmente otimistas. Em materiais, vazamentos comuns incluem o uso de múltiplas medições correlacionadas do mesmo espécime em conjuntos diferentes, ou a extração de estatísticas globais do conjunto completo antes da separação.

A regularização é o conjunto de técnicas para controlar a complexidade efetiva do modelo, reduzindo overfitting. Overfitting ocorre quando o modelo aprende padrões específicos do conjunto de treino, incluindo ruído experimental, e perde capacidade preditiva em dados novos. Underfitting ocorre quando o modelo é simples demais e não captura relações relevantes. O equilíbrio entre viés e variância expressa esse compromisso: modelos mais simples tendem a alto viés e baixa variância, enquanto modelos mais complexos tendem a baixo viés e alta variância. A escolha apropriada depende do tamanho do conjunto de dados, do nível de ruído e da complexidade do fenômeno físico modelado.

A representação dos dados é frequentemente o fator mais determinante do sucesso. Para dados tabulares, a engenharia de atributos envolve selecionar variáveis, tratar escalas, lidar com valores ausentes e codificar variáveis categóricas. Para imagens microestruturais, a representação pode ser feita por descritores clássicos, como texturas e granulometria, ou aprendida automaticamente por redes convolucionais. Para séries temporais, como sinais de ensaio, pode-se extrair estatísticas e parâmetros físicos, ou usar modelos capazes de lidar com sequências. Em todos os casos, a escolha de atributos deve refletir hipóteses plausíveis sobre mecanismos materiais, de modo que o modelo aprenda relações significativas e não apenas correlações espúrias.

Interpretabilidade e explicabilidade são requisitos frequentes em ciência dos materiais. Mesmo quando o objetivo principal é previsão, é comum desejar entender quais variáveis mais contribuem para uma propriedade e como mudanças em composição ou processamento alteram o resultado. Alguns modelos, como regressões e árvores, são mais transparentes. Outros, como ensembles complexos e redes profundas, exigem técnicas adicionais de explicação, como análise de importância de atributos, permutação, SHAP ou gradientes em modelos de imagem. A interpretação deve ser tratada com cautela, pois relações estatísticas não implicam causalidade. Em materiais, correlações podem refletir co-variações indiretas, como um parâmetro de processamento que muda simultaneamente múltiplos aspectos microestruturais.

Por fim, a prática de ML em materiais demanda atenção a qualidade e rastreabilidade dos dados. Variabilidade de laboratório, diferenças de protocolo, condições ambientais e calibração de instrumentos podem introduzir vieses. A curadoria do dataset, a padronização de unidades, a documentação de origem e a validação física dos resultados são parte do processo. Um modelo com alto desempenho numérico, mas inconsistente com princípios fundamentais, pode ser inútil ou perigoso em contexto de engenharia. Assim, a formulação do problema, a escolha de métricas e a validação devem sempre ser guiadas por conhecimento de domínio, garantindo que o aprendizado de máquina opere como ferramenta científica e não como substituto acrítico da compreensão material.

## 3. Tipos de dados em ciência dos materiais

Os problemas de aprendizado de máquina em ciência dos materiais diferenciam-se principalmente pela natureza e pela heterogeneidade dos dados envolvidos. De forma geral, esses dados podem ser classificados em dados tabulares, dados de sinais, dados de imagens e dados derivados de simulações computacionais. A compreensão do tipo de dado é essencial para a escolha adequada do modelo, das técnicas de pré-processamento e das métricas de avaliação.

Dados tabulares são os mais comuns e estruturados, organizados em linhas e colunas. Cada linha representa um material, amostra ou condição experimental, enquanto as colunas representam atributos como composição química, parâmetros de processamento, condições de ensaio ou propriedades medidas. Esse tipo de dado é amplamente utilizado na previsão de propriedades mecânicas, térmicas, elétricas e químicas, sendo compatível com a maioria dos algoritmos clássicos de aprendizado supervisionado e não supervisionado.

Dados de sinais surgem em ensaios experimentais que produzem respostas contínuas no tempo ou em função de outra variável controlada. Exemplos incluem curvas tensão deformação, sinais de fadiga, respostas dinâmicas, espectros de difração de raios X e espectros de infravermelho ou Raman. Esses dados podem ser utilizados diretamente como séries ou convertidos em descritores físicos e estatísticos, permitindo o uso de modelos de regressão, classificação ou técnicas de redução de dimensionalidade.

Dados de imagens são centrais em aplicações modernas de ciência dos materiais. Imagens de microscopia óptica, eletrônica de varredura ou transmissão, tomografia e inspeção visual de superfícies carregam informações ricas sobre microestrutura, defeitos, fases e morfologia. Essas imagens podem ser analisadas por descritores manuais, como tamanho de grão e textura, ou diretamente por modelos de aprendizado profundo, especialmente redes convolucionais, que aprendem representações relevantes de forma automática.

Dados provenientes de simulações computacionais incluem resultados de métodos como dinâmica molecular, elementos finitos, primeiros princípios e simulações termodinâmicas. Esses dados podem assumir forma tabular, vetorial ou mesmo campos espaciais e temporais. O aprendizado de máquina pode ser utilizado para acelerar simulações, aproximar modelos físicos complexos ou explorar grandes espaços de parâmetros com custo computacional reduzido.

Na prática, muitos problemas reais em materiais envolvem a combinação de diferentes tipos de dados, como atributos tabulares associados a imagens microestruturais ou sinais experimentais. Essa multimodalidade impõe desafios adicionais de integração, normalização e interpretação, mas também amplia o potencial do aprendizado de máquina para capturar relações mais completas entre processamento, estrutura e propriedades.

| Tipo de dado | Forma típica | Origem comum | Uso em aprendizado de máquina em materiais |

| Dados de simulação | Vetores, campos ou séries | Dinâmica molecular, elementos finitos, primeiros princípios | Aceleração de simulações, aproximação de modelos físicos e exploração de espaço de parâmetros |

## 4. Fluxo geral de um problema de aprendizado de máquina em ciência dos materiais

A aplicação de aprendizado de máquina em ciência dos materiais deve ser encarada como um processo estruturado, no qual decisões metodológicas têm impacto direto na validade científica dos resultados. O primeiro passo é a formulação clara do problema, definindo o objetivo do modelo, o tipo de saída desejada e o papel que o modelo terá no contexto experimental ou computacional. Essa etapa exige conhecimento do fenômeno material envolvido, pois a definição inadequada do problema conduz a modelos estatisticamente corretos, porém cientificamente irrelevantes.

Em seguida, ocorre a coleta e organização dos dados. Dados experimentais e de simulação devem ser reunidos de forma consistente, com atenção à padronização de unidades, condições de ensaio e protocolos de medição. A rastreabilidade das amostras e a documentação das condições de obtenção dos dados são fundamentais para evitar ambiguidades e vieses ocultos. Nesta fase, já se avalia a quantidade de dados disponível, sua variabilidade e a presença de ruído experimental.

O pré-processamento constitui uma etapa central do fluxo. Inclui limpeza de dados, tratamento de valores ausentes, remoção de inconsistências e normalização ou padronização das variáveis. Também envolve a seleção e a construção de atributos relevantes, processo conhecido como engenharia de atributos. Em ciência dos materiais, essa etapa deve ser orientada por princípios físicos e metalúrgicos, evitando o uso acrítico de variáveis que não tenham significado ou que introduzam correlações espúrias.

Após o pré-processamento, os dados são divididos em conjuntos distintos para treinamento, validação e teste. Essa separação permite avaliar a capacidade de generalização do modelo e reduz o risco de conclusões artificiais. Em bases pequenas, técnicas como validação cruzada são preferíveis. Deve-se garantir que amostras correlacionadas não sejam distribuídas entre conjuntos diferentes, preservando a independência estatística das avaliações.

O treinamento do modelo envolve a escolha do algoritmo, o ajuste de seus parâmetros internos e a definição de hiperparâmetros. Essa escolha deve considerar o tipo de dado, o tamanho da base, o nível de ruído e a necessidade de interpretabilidade. Modelos mais simples podem ser preferíveis em contextos com poucos dados ou quando a explicação física é prioritária, enquanto modelos mais complexos podem ser justificados em problemas altamente não lineares e com grande volume de dados.

A avaliação do desempenho deve ir além de métricas numéricas isoladas. É necessário analisar a estabilidade do modelo, sua sensibilidade a variações nos dados e a coerência dos resultados com o conhecimento material existente. Resultados inconsistentes com leis físicas conhecidas indicam problemas de formulação, dados ou modelo. A interpretação dos resultados deve ser feita com cautela, reconhecendo que aprendizado de máquina identifica correlações estatísticas e não estabelece causalidade por si só.

Por fim, a etapa de uso e validação científica envolve integrar o modelo ao fluxo de pesquisa ou desenvolvimento. Isso pode incluir a geração de hipóteses, a priorização de experimentos, a otimização de parâmetros ou o suporte à tomada de decisão. O modelo deve ser continuamente reavaliado à medida que novos dados se tornam disponíveis, mantendo-se como uma ferramenta dinâmica e integrada ao processo científico, e não como uma solução definitiva ou isolada.

## 5. Regressão linear e regressão polinomial em materiais

Regressão é apropriada quando o objetivo é prever uma propriedade contínua. Em ciência dos materiais, isso ocorre quando se deseja estimar valores como dureza, limite de escoamento, resistência à tração, condutividade ou vida em fadiga a partir de variáveis de composição, processamento e microestrutura. A regressão linear assume que a propriedade pode ser descrita como combinação linear dos atributos. Essa hipótese é útil como linha de base por sua simplicidade e interpretabilidade, porém pode falhar quando a relação é não linear. A regressão polinomial estende o modelo linear ao incluir termos de potência e interação, permitindo capturar curvaturas e efeitos combinados frequentemente presentes em materiais, mantendo a mesma ideia de ajuste por mínimos quadrados.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import Pipeline
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# =========================================================
# 1. Geração de dados sintéticos inspirados em materiais
# =========================================================

rng = np.random.default_rng(42)
n = 220

# Variáveis de entrada
C = rng.uniform(0.05, 0.9, n)       # teor de carbono (%)
Mn = rng.uniform(0.2, 1.8, n)       # teor de manganês (%)
T_rev = rng.uniform(150, 650, n)    # temperatura de revenimento (°C)

# Relação sintética para dureza (HV)
noise = rng.normal(0, 18, n)
HV = (
    180
    + 520 * C
    + 55 * Mn
    - 0.28 * T_rev
    + 260 * (C ** 2)
    - 0.015 * (T_rev * C)
    + noise
)

df = pd.DataFrame({
    "C_pct": C,
    "Mn_pct": Mn,
    "T_rev_C": T_rev,
    "HV": HV
})

X = df[["C_pct", "Mn_pct", "T_rev_C"]]
y = df["HV"]

# =========================================================
# 2. Divisão treino e teste
# =========================================================

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=7
)

# =========================================================
# 3. Modelos de regressão
# =========================================================

# Regressão linear
lin = LinearRegression()
lin.fit(X_train, y_train)
y_pred_lin = lin.predict(X_test)

# Regressão polinomial (grau 2)
poly2 = Pipeline([
    ("poly", PolynomialFeatures(degree=2, include_bias=False)),
    ("lr", LinearRegression())
])
poly2.fit(X_train, y_train)
y_pred_poly = poly2.predict(X_test)

# =========================================================
# 4. Métricas
# =========================================================

def regression_metrics(y_true, y_pred):
    mae = mean_absolute_error(y_true, y_pred)
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_true, y_pred)
    return mae, rmse, r2

metrics_lin = regression_metrics(y_test, y_pred_lin)
metrics_poly = regression_metrics(y_test, y_pred_poly)

metrics_df = pd.DataFrame(
    [
        ("Regressão linear", *metrics_lin),
        ("Regressão polinomial (grau 2)", *metrics_poly),
    ],
    columns=["Modelo", "MAE (HV)", "RMSE (HV)", "R²"]
)

print(metrics_df)

# =========================================================
# 5. Gráficos: valores reais vs preditos
# =========================================================

plt.figure()
plt.scatter(y_test, y_pred_lin)
plt.plot([y_test.min(), y_test.max()],
         [y_test.min(), y_test.max()])
plt.xlabel("Dureza real (HV)")
plt.ylabel("Dureza predita (HV)")
plt.title("Regressão linear: valores reais vs preditos")
plt.show()

plt.figure()
plt.scatter(y_test, y_pred_poly)
plt.plot([y_test.min(), y_test.max()],
         [y_test.min(), y_test.max()])
plt.xlabel("Dureza real (HV)")
plt.ylabel("Dureza predita (HV)")
plt.title("Regressão polinomial (grau 2): valores reais vs preditos")
plt.show()

# =========================================================
# 6. Gráficos de resíduos
# =========================================================

res_lin = y_test - y_pred_lin
res_poly = y_test - y_pred_poly

plt.figure()
plt.scatter(y_pred_lin, res_lin)
plt.axhline(0)
plt.xlabel("Dureza predita (HV)")
plt.ylabel("Resíduo (HV)")
plt.title("Resíduos: regressão linear")
plt.show()

plt.figure()
plt.scatter(y_pred_poly, res_poly)
plt.axhline(0)
plt.xlabel("Dureza predita (HV)")
plt.ylabel("Resíduo (HV)")
plt.title("Resíduos: regressão polinomial (grau 2)")
plt.show()
```

A implementação apresentada representa um fluxo completo e coerente de regressão aplicado a um problema típico da ciência dos materiais, no qual uma propriedade contínua é influenciada simultaneamente por composição química e parâmetros de processamento. A dureza foi escolhida como variável alvo por ser uma propriedade amplamente utilizada como indicador indireto de microestrutura e estado metalúrgico, enquanto teor de carbono, teor de manganês e temperatura de revenimento representam variáveis fisicamente relevantes e com significado metalúrgico claro.

O modelo de regressão linear atua como referência conceitual e prática. Seus coeficientes permitem interpretação direta, indicando tendências médias de influência de cada variável quando as demais são mantidas constantes. Esse tipo de modelo é particularmente útil em materiais quando o objetivo é análise exploratória, identificação de variáveis dominantes ou construção de modelos simplificados coerentes com hipóteses físicas iniciais. No entanto, sua limitação está na incapacidade de representar curvaturas e efeitos de interação comuns em fenômenos termoquímicos e microestruturais.

A regressão polinomial de segundo grau amplia essa capacidade ao introduzir termos quadráticos e cruzados entre variáveis. Esses termos permitem modelar, por exemplo, respostas não lineares da dureza ao teor de carbono e a dependência do efeito do revenimento em função da composição. Embora matematicamente mais complexa, essa abordagem ainda preserva a estrutura de um modelo linear nos parâmetros, mantendo estabilidade numérica e interpretação possível, especialmente quando guiada por conhecimento de domínio.

As métricas de erro quantificam o desempenho preditivo em unidades fisicamente significativas. O erro absoluto médio fornece uma noção direta do desvio médio esperado em dureza, enquanto o RMSE penaliza discrepâncias maiores, refletindo sensibilidade a erros extremos. O coeficiente de determinação indica o quanto da variabilidade experimental é explicada pelo modelo, mas não deve ser analisado isoladamente.

A análise gráfica complementa as métricas numéricas. Os gráficos de valores reais versus preditos permitem verificar a proximidade dos pontos em relação à diagonal ideal, indicando qualidade global do ajuste. Já os gráficos de resíduos são fundamentais para identificar padrões sistemáticos, como curvaturas não capturadas, variação dependente da magnitude da predição ou indícios de underfitting. Em ciência dos materiais, a presença de padrões estruturados nos resíduos geralmente indica fenômenos físicos relevantes não representados no modelo ou a necessidade de atributos adicionais.

Em conjunto, esse exemplo ilustra não apenas a aplicação técnica de regressão, mas a forma correta de raciocinar sobre modelos preditivos em materiais. O modelo deve ser avaliado simultaneamente sob critérios estatísticos, visuais e físicos, garantindo que a capacidade de previsão esteja alinhada com a interpretação científica e com o entendimento dos mecanismos materiais envolvidos.

## 6. Regressão baseada em árvores em ciência dos materiais

Modelos baseados em árvores de decisão são amplamente utilizados em ciência dos materiais por sua capacidade de representar relações não lineares sem exigir transformações explícitas das variáveis de entrada. Diferentemente da regressão linear, esses modelos particionam o espaço de atributos em regiões, tomando decisões sucessivas baseadas em limiares das variáveis. Cada partição corresponde a um regime distinto do comportamento do material, o que se alinha bem com fenômenos como mudanças microestruturais, transições de fase ou efeitos de processamento.

A árvore de decisão para regressão constrói regras do tipo “se-então”, aprendidas diretamente dos dados, permitindo interpretar quais variáveis e quais faixas são mais relevantes para a previsão de uma propriedade. Entretanto, árvores individuais tendem a ser instáveis e sensíveis a ruído experimental. Para contornar essa limitação, métodos de conjunto, como Random Forest, combinam múltiplas árvores treinadas sobre subconjuntos dos dados, produzindo modelos mais robustos e com melhor capacidade de generalização, característica valiosa em bases experimentais heterogêneas de materiais.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# =========================================================
# 1. Dados sintéticos inspirados em ciência dos materiais
# Objetivo: prever limite de escoamento (MPa)
# Entradas: tamanho de grão, %C e temperatura de tratamento
# =========================================================

rng = np.random.default_rng(10)
n = 240

grain_size = rng.uniform(2, 50, n)     # tamanho de grão (µm)
C = rng.uniform(0.05, 0.9, n)           # teor de carbono (%)
T_ht = rng.uniform(750, 1050, n)        # temperatura de tratamento térmico (°C)

noise = rng.normal(0, 22, n)

# Relação sintética com não linearidades e limiares
sigma_y = (
    120
    + 380 / np.sqrt(grain_size)         # efeito Hall-Petch
    + 450 * C
    - 0.15 * T_ht
    + noise
)

df = pd.DataFrame({
    "grain_size_um": grain_size,
    "C_pct": C,
    "T_ht_C": T_ht,
    "sigma_y_MPa": sigma_y
})

X = df[["grain_size_um", "C_pct", "T_ht_C"]]
y = df["sigma_y_MPa"]

# =========================================================
# 2. Divisão treino e teste
# =========================================================

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=4
)

# =========================================================
# 3. Modelos baseados em árvores
# =========================================================

# Árvore de decisão
tree = DecisionTreeRegressor(max_depth=4, random_state=4)
tree.fit(X_train, y_train)
y_pred_tree = tree.predict(X_test)

# Random Forest
rf = RandomForestRegressor(
    n_estimators=200,
    max_depth=6,
    random_state=4
)
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)

# =========================================================
# 4. Métricas
# =========================================================

def metrics(y_true, y_pred):
    mae = mean_absolute_error(y_true, y_pred)
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    r2 = r2_score(y_true, y_pred)
    return mae, rmse, r2

m_tree = metrics(y_test, y_pred_tree)
m_rf = metrics(y_test, y_pred_rf)

metrics_df = pd.DataFrame(
    [
        ("Árvore de decisão", *m_tree),
        ("Random Forest", *m_rf),
    ],
    columns=["Modelo", "MAE (MPa)", "RMSE (MPa)", "R²"]
)

print(metrics_df)

# =========================================================
# 5. Gráficos: valores reais vs preditos
# =========================================================

plt.figure()
plt.scatter(y_test, y_pred_tree)
plt.plot([y_test.min(), y_test.max()],
         [y_test.min(), y_test.max()])
plt.xlabel("Limite de escoamento real (MPa)")
plt.ylabel("Limite de escoamento predito (MPa)")
plt.title("Árvore de decisão: valores reais vs preditos")
plt.show()

plt.figure()
plt.scatter(y_test, y_pred_rf)
plt.plot([y_test.min(), y_test.max()],
         [y_test.min(), y_test.max()])
plt.xlabel("Limite de escoamento real (MPa)")
plt.ylabel("Limite de escoamento predito (MPa)")
plt.title("Random Forest: valores reais vs preditos")
plt.show()

# =========================================================
# 6. Importância das variáveis (Random Forest)
# =========================================================

importances = pd.Series(
    rf.feature_importances_,
    index=X.columns
).sort_values(ascending=False)

print("\nImportância das variáveis (Random Forest):")
print(importances.round(3))
```

Neste exemplo, o limite de escoamento é modelado como função do tamanho de grão, da composição química e da temperatura de tratamento térmico, refletindo um cenário físico realista. A árvore de decisão constrói regras explícitas que particionam o espaço de variáveis, aproximando regimes distintos de comportamento mecânico. Essa característica permite interpretar o modelo como uma sequência de decisões associadas a faixas microestruturais e processuais.

O Random Forest combina múltiplas árvores para reduzir a sensibilidade a ruído e a instabilidades típicas de dados experimentais. O ganho de desempenho observado em relação à árvore única reflete a maior robustez estatística do conjunto. A análise de importância das variáveis fornece uma indicação quantitativa de quais fatores mais influenciam a propriedade prevista, sendo particularmente útil para orientar experimentos, reduzir dimensionalidade e verificar coerência com princípios conhecidos, como o efeito Hall-Petch e o papel da composição química.

Esse tipo de modelo é especialmente adequado em ciência dos materiais quando relações não lineares são esperadas, quando há interações complexas entre variáveis e quando a interpretabilidade local ou global do modelo é desejável sem recorrer a arquiteturas profundas.

## 7. Aprendizado não supervisionado e clustering em ciência dos materiais

Em muitos problemas de ciência dos materiais não existem rótulos prévios ou valores alvo claramente definidos. Nessas situações, o aprendizado não supervisionado é utilizado para explorar a estrutura interna dos dados, identificar padrões e agrupar amostras com comportamento semelhante. O clustering é particularmente útil para analisar grandes conjuntos experimentais, nos quais diferentes ligas, tratamentos térmicos ou estados microestruturais coexistem sem classificação explícita.

O algoritmo k-means é um dos métodos mais utilizados para esse fim. Ele particiona os dados em um número pré-definido de grupos, minimizando a variabilidade interna de cada grupo e maximizando a separação entre eles. Em ciência dos materiais, esses grupos podem ser interpretados como famílias de materiais com propriedades mecânicas semelhantes, regimes de comportamento distintos ou classes microestruturais emergentes, auxiliando tanto a análise exploratória quanto a formulação de hipóteses físicas.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# =========================================================
# 1. Dados sintéticos inspirados em ciência dos materiais
# Objetivo: agrupar ligas por comportamento mecânico
# =========================================================

rng = np.random.default_rng(21)
n = 180

# Propriedades mecânicas medidas
E = rng.normal(210, 15, n)          # módulo de elasticidade (GPa)
sigma_y = rng.normal(420, 60, n)    # limite de escoamento (MPa)
sigma_u = rng.normal(620, 80, n)    # resistência à tração (MPa)
elong = rng.normal(18, 5, n)        # alongamento (%)

# Introdução de grupos artificiais representando ligas distintas
E[:60] += 20
sigma_y[:60] += 120

elong[60:120] += 10
sigma_u[60:120] -= 60

df = pd.DataFrame({
    "E_GPa": E,
    "sigma_y_MPa": sigma_y,
    "sigma_u_MPa": sigma_u,
    "elong_pct": elong
})

# =========================================================
# 2. Padronização dos dados
# =========================================================

scaler = StandardScaler()
X_scaled = scaler.fit_transform(df)

# =========================================================
# 3. Clustering com k-means
# =========================================================

kmeans = KMeans(n_clusters=3, random_state=21, n_init=20)
clusters = kmeans.fit_predict(X_scaled)

df["cluster"] = clusters

# =========================================================
# 4. Redução de dimensionalidade para visualização
# =========================================================

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

df["PC1"] = X_pca[:, 0]
df["PC2"] = X_pca[:, 1]

# =========================================================
# 5. Visualização dos clusters
# =========================================================

plt.figure()
plt.scatter(df["PC1"], df["PC2"], c=df["cluster"])
plt.xlabel("Componente principal 1")
plt.ylabel("Componente principal 2")
plt.title("Clustering de ligas por propriedades mecânicas")
plt.show()

# =========================================================
# 6. Estatísticas médias por cluster
# =========================================================

cluster_summary = df.groupby("cluster").mean()
cluster_summary
```

O código exemplifica o uso de clustering para explorar dados de propriedades mecânicas sem qualquer rótulo prévio. As variáveis são inicialmente padronizadas para garantir que diferenças de escala não dominem o processo de agrupamento, passo essencial quando propriedades possuem unidades e magnitudes distintas.

O k-means identifica grupos de amostras com comportamento mecânico semelhante, que podem ser interpretados como diferentes famílias de ligas ou estados resultantes de processamento distinto. A redução de dimensionalidade por PCA permite visualizar esses agrupamentos em duas dimensões, facilitando a análise qualitativa da separação entre grupos.

As médias de cada cluster fornecem subsídios para interpretação física, permitindo associar cada grupo a níveis típicos de rigidez, resistência e ductilidade. Em ciência dos materiais, esse tipo de análise é frequentemente utilizado para organizar grandes bases experimentais, identificar regimes de comportamento e orientar estudos mais aprofundados, mesmo antes da definição de um modelo preditivo supervisionado.

## 8. Classificação supervisionada aplicada à identificação de estados e defeitos em materiais

Problemas de classificação surgem quando o objetivo é atribuir uma amostra a uma classe discreta previamente definida. Em ciência dos materiais, isso ocorre, por exemplo, na identificação de presença ou ausência de defeitos, classificação de modos de fratura, distinção entre estados microestruturais ou separação de regimes de comportamento mecânico. Diferentemente da regressão, o foco não está em prever um valor contínuo, mas em aprender fronteiras de decisão que separem classes com base em atributos físicos mensuráveis.

Modelos simples de classificação, como Regressão Logística e k-Nearest Neighbors, são frequentemente utilizados como ponto de partida por sua interpretação direta e baixo custo computacional. Eles permitem avaliar se as variáveis disponíveis contêm informação suficiente para discriminar classes de interesse, antes de recorrer a modelos mais complexos. Em materiais, essa abordagem é particularmente útil quando se deseja justificar decisões experimentais ou inspeções automáticas com base em critérios mensuráveis.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay

# =========================================================
# 1. Dados sintéticos inspirados em ciência dos materiais
# Objetivo: classificar defeito vs sem defeito
# =========================================================

rng = np.random.default_rng(33)
n = 200

porosity = rng.uniform(0, 5, n)            # porosidade (%)
grain_size = rng.uniform(5, 40, n)         # tamanho de grão (µm)
sigma_u = rng.normal(650, 70, n)           # resistência à tração (MPa)

# Regra física simplificada para defeito
defect = ((porosity > 2.5) | (sigma_u < 600)).astype(int)

df = pd.DataFrame({
    "porosity_pct": porosity,
    "grain_size_um": grain_size,
    "sigma_u_MPa": sigma_u,
    "defect": defect
})

# =========================================================
# 2. Classificação supervisionada (3 variáveis)
# =========================================================

X = df[["porosity_pct", "grain_size_um", "sigma_u_MPa"]]
y = df["defect"]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=12, stratify=y
)

scaler = StandardScaler()
X_train_sc = scaler.fit_transform(X_train)
X_test_sc = scaler.transform(X_test)

clf = LogisticRegression()
clf.fit(X_train_sc, y_train)
y_pred = clf.predict(X_test_sc)

print(f"Acurácia: {accuracy_score(y_test, y_pred):.3f}")

ConfusionMatrixDisplay(
    confusion_matrix(y_test, y_pred),
    display_labels=["Sem defeito", "Defeito"]
).plot(values_format="d")
plt.title("Matriz de confusão – regressão logística")
plt.show()

# =========================================================
# 3. Região de decisão (visualização em 2D)
# Usando porosidade e resistência à tração
# =========================================================

X2 = df[["porosity_pct", "sigma_u_MPa"]].values
y2 = df["defect"].values

scaler2 = StandardScaler()
X2_sc = scaler2.fit_transform(X2)

clf2 = LogisticRegression()
clf2.fit(X2_sc, y2)

# Malha de pontos
x_min, x_max = X2[:, 0].min() - 0.5, X2[:, 0].max() + 0.5
y_min, y_max = X2[:, 1].min() - 20, X2[:, 1].max() + 20

xx, yy = np.meshgrid(
    np.linspace(x_min, x_max, 300),
    np.linspace(y_min, y_max, 300)
)

grid = np.c_[xx.ravel(), yy.ravel()]
grid_sc = scaler2.transform(grid)
Z = clf2.predict(grid_sc).reshape(xx.shape)

# Plot da região de decisão
plt.figure()
plt.contourf(xx, yy, Z, alpha=0.3)
plt.scatter(
    X2[y2 == 0, 0], X2[y2 == 0, 1], label="Sem defeito"
)
plt.scatter(
    X2[y2 == 1, 0], X2[y2 == 1, 1], label="Defeito"
)
plt.xlabel("Porosidade (%)")
plt.ylabel("Resistência à tração (MPa)")
plt.title("Região de decisão – regressão logística")
plt.legend()
plt.show()
```

A análise realizada ilustra de forma completa como um classificador supervisionado deve ser interpretado em ciência dos materiais, indo além do simples cálculo de métricas. O ponto central não é apenas saber se o modelo acerta, mas compreender como e por que ele toma determinadas decisões a partir de variáveis fisicamente significativas.

No primeiro nível, o classificador é treinado com três atributos que representam mecanismos distintos de degradação mecânica: porosidade, tamanho de grão e resistência à tração. A presença de defeitos é tratada como um problema de decisão binária, típico de inspeção de qualidade e avaliação de confiabilidade estrutural. A acurácia e a matriz de confusão fornecem uma visão global do desempenho, permitindo identificar a taxa de acertos e os tipos de erro cometidos. Em materiais, essa distinção é crítica, pois falsos negativos, isto é, classificar um material defeituoso como adequado, têm impacto direto na segurança e no desempenho do sistema.

No segundo nível, a análise é deliberadamente reduzida a duas variáveis para permitir a visualização da região de decisão. Essa redução não tem objetivo operacional, mas interpretativo. Ao projetar o problema no plano definido por porosidade e resistência à tração, torna-se possível observar explicitamente a fronteira aprendida pelo modelo. A regressão logística produz uma fronteira linear, o que significa que o modelo está aprendendo um critério de separação baseado em combinações ponderadas dessas variáveis, equivalente a um limiar técnico contínuo.

A região de decisão revela como o classificador internaliza relações físicas conhecidas. Áreas do espaço com alta porosidade e baixa resistência são associadas à classe defeituosa, enquanto regiões com baixa porosidade e resistência elevada são associadas à classe sem defeito. A inclinação da fronteira indica que não existe um único limite rígido para cada variável isoladamente, mas sim uma relação de compromisso entre elas. Esse comportamento é coerente com a prática experimental, na qual materiais com porosidade moderada podem ser aceitáveis se a resistência for suficientemente alta, e vice-versa.

Essa visualização permite identificar zonas críticas próximas à fronteira de decisão, onde pequenas variações experimentais podem alterar a classificação. Em ciência dos materiais, essas regiões são particularmente importantes, pois correspondem a estados limítrofes de processamento ou microestrutura, nos quais a variabilidade experimental, o ruído de medição ou a heterogeneidade do material podem ter grande impacto na confiabilidade da decisão. A análise da região de decisão, portanto, auxilia na definição de margens de segurança, critérios de aceitação e necessidade de ensaios adicionais.

Por fim, esse exemplo evidencia uma prática essencial no uso de aprendizado de máquina em materiais: modelos de classificação não devem ser tratados como caixas-pretas isoladas. A interpretação geométrica da decisão, combinada com métricas quantitativas e conhecimento físico, é o que confere validade científica e utilidade prática ao modelo. A região de decisão funciona como uma ponte entre o formalismo matemático do classificador e a compreensão material do fenômeno analisado, permitindo avaliar se o modelo aprende padrões reais ou apenas ajusta correlações estatísticas sem significado físico.

## 9. Análise e classificação de imagens reais de defeitos superficiais em materiais

A estrutura do dataset está correta e agora conhecida. Trata-se do **NEU-DET**, um conjunto real de imagens de superfícies metálicas com seis classes de defeitos. As imagens já estão separadas em conjuntos de treino e validação, organizadas exatamente no formato exigido por pipelines de visão computacional baseados em CNNs. A presença de arquivos XML indica anotações adicionais, mas para classificação pura de imagens elas não são utilizadas.

O objetivo aqui é realizar **análise automática das imagens** por meio de uma rede neural convolucional, aprendendo padrões visuais associados a diferentes mecanismos físicos de defeito superficial, como trincas, inclusões, riscos e irregularidades de laminação.


```python
import os
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf

from tensorflow.keras import layers, models
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# =========================================================
# 1. Diretórios corretos (detectados automaticamente)
# =========================================================

train_dir = "dataset_inspecionado/NEU-DET/train/images"
val_dir = "dataset_inspecionado/NEU-DET/validation/images"

print("Classes disponíveis:", os.listdir(train_dir))

# =========================================================
# 2. Preparação dos dados
# =========================================================

img_size = (128, 128)
batch_size = 32

train_datagen = ImageDataGenerator(
    rescale=1.0 / 255.0,
    rotation_range=10,
    width_shift_range=0.05,
    height_shift_range=0.05,
    zoom_range=0.1,
    horizontal_flip=True
)

val_datagen = ImageDataGenerator(
    rescale=1.0 / 255.0
)

train_gen = train_datagen.flow_from_directory(
    train_dir,
    target_size=img_size,
    batch_size=batch_size,
    class_mode="categorical"
)

val_gen = val_datagen.flow_from_directory(
    val_dir,
    target_size=img_size,
    batch_size=batch_size,
    class_mode="categorical",
    shuffle=False
)

num_classes = train_gen.num_classes
print("Número de classes:", num_classes)
print("Mapeamento de classes:", train_gen.class_indices)

# =========================================================
# 3. Visualização de amostras
# =========================================================

images, labels = next(train_gen)

plt.figure(figsize=(8, 8))
for i in range(9):
    plt.subplot(3, 3, i + 1)
    plt.imshow(images[i])
    plt.axis("off")
plt.suptitle("Amostras de imagens de defeitos superficiais")
plt.show()

# =========================================================
# 4. Definição da CNN
# =========================================================

model = models.Sequential([
    layers.Conv2D(32, (3, 3), activation="relu",
                  input_shape=(128, 128, 3)),
    layers.MaxPooling2D(2),

    layers.Conv2D(64, (3, 3), activation="relu"),
    layers.MaxPooling2D(2),

    layers.Conv2D(128, (3, 3), activation="relu"),
    layers.MaxPooling2D(2),

    layers.Flatten(),
    layers.Dense(256, activation="relu"),
    layers.Dropout(0.4),
    layers.Dense(num_classes, activation="softmax")
])

model.compile(
    optimizer="adam",
    loss="categorical_crossentropy",
    metrics=["accuracy"]
)

model.summary()

# =========================================================
# 5. Treinamento
# =========================================================

history = model.fit(
    train_gen,
    validation_data=val_gen,
    epochs=10
)

# =========================================================
# 6. Curvas de aprendizado
# =========================================================

plt.figure()
plt.plot(history.history["accuracy"], label="Treino")
plt.plot(history.history["val_accuracy"], label="Validação")
plt.xlabel("Época")
plt.ylabel("Acurácia")
plt.title("CNN – Classificação de defeitos superficiais (NEU-DET)")
plt.legend()
plt.show()
```

A análise realizada utiliza diretamente imagens reais de defeitos superficiais em chapas metálicas, organizadas em classes que correspondem a diferentes mecanismos físicos de dano. A separação prévia entre treino e validação elimina a necessidade de divisão manual e permite avaliar generalização de forma direta.

O pré-processamento inclui normalização e pequenas transformações geométricas, simulando variações experimentais comuns, como mudanças de posição, orientação e iluminação. Essas operações aumentam a robustez do modelo frente à variabilidade típica de dados industriais reais.

A rede neural convolucional aprende hierarquicamente padrões visuais. As primeiras camadas extraem bordas, texturas e contrastes locais. Camadas intermediárias capturam padrões mais complexos associados à morfologia dos defeitos, como trincas alongadas, regiões pontilhadas ou áreas irregulares. As camadas finais realizam a decisão multiclasse, associando cada imagem a um tipo específico de defeito.

As curvas de aprendizado permitem discutir aspectos centrais do uso de aprendizado profundo em ciência dos materiais, como convergência, overfitting e limitação de dados. Esse exemplo representa fielmente aplicações reais de inspeção automática, controle de qualidade e análise microestrutural assistida por inteligência artificial, encerrando o ciclo de técnicas de aprendizado de máquina aplicadas a dados tabulares, sinais experimentais e imagens de materiais.

O coeficiente de determinação pode indicar o quanto da variabilidade dos dados é explicada pelo modelo, mas não deve ser usado isoladamente, pois valores elevados podem coexistir com erros sistemáticos relevantes. Por isso, a análise gráfica dos resultados é indispensável. Gráficos de valores reais versus preditos permitem avaliar a qualidade global do ajuste, enquanto gráficos de resíduos revelam padrões estruturados que indicam não linearidades não modeladas, efeitos de saturação, mudança de regime físico ou dependência do erro com a magnitude da propriedade. Em ciência dos materiais, esses padrões frequentemente correspondem a transições microestruturais, limiares de processamento ou mecanismos concorrentes que não foram explicitamente representados nos atributos de entrada.

Em problemas de classificação, a avaliação exige ainda mais cuidado. Métricas agregadas como acurácia escondem assimetrias críticas entre tipos de erro. A matriz de confusão é fundamental para distinguir falsos positivos e falsos negativos e interpretar suas implicações físicas e industriais. Em inspeção de materiais, classificar um material defeituoso como adequado pode levar a falhas em serviço, enquanto rejeitar um material aceitável implica custo econômico. Portanto, a escolha do modelo e do limiar de decisão deve ser orientada pelo contexto de uso, e não apenas pelo desempenho médio. A visualização da região de decisão, quando possível, complementa essa análise ao revelar como o modelo organiza o espaço das variáveis físicas. Regiões próximas à fronteira de decisão indicam estados limítrofes, nos quais pequenas variações experimentais, ruído de medição ou heterogeneidade microestrutural podem alterar a classificação. Essas regiões são particularmente relevantes em materiais, pois frequentemente correspondem a condições críticas de processamento ou serviço. Em tarefas com imagens, a avaliação deve considerar não apenas métricas globais, mas também a inspeção qualitativa de erros por classe, identificando confusões recorrentes entre defeitos com morfologia semelhante, o que pode indicar necessidade de mais dados, maior resolução ou ajustes na aquisição experimental.

Do ponto de vista prático, o uso de aprendizado de máquina em ciência dos materiais é fortemente condicionado por restrições de dados, qualidade experimental e coerência física. Bases de dados são tipicamente pequenas, caras e difíceis de expandir, pois dependem de ensaios destrutivos, instrumentação sofisticada e tempo de laboratório. Isso impõe um risco elevado de overfitting, especialmente em modelos flexíveis, e torna essencial o controle de complexidade, a validação adequada e a preferência por modelos mais simples quando os dados não sustentam maior capacidade expressiva. A qualidade dos dados é determinante: inconsistências de protocolo, variações de operador, diferenças de lote, deriva de calibração e condições ambientais introduzem vieses que o modelo aprende como se fossem relações físicas reais. Por esse motivo, a curadoria dos dados, a padronização de unidades, a documentação de condições experimentais e a rastreabilidade das amostras fazem parte do processo de modelagem e não são etapas externas ao aprendizado de máquina.

Outro ponto crítico é a interpretação dos resultados. Modelos de aprendizado de máquina identificam correlações estatísticas, não causalidade. Em sistemas materiais complexos, correlações podem emergir de variáveis intermediárias, efeitos de confusão ou co-variações indiretas entre composição, processamento e microestrutura. A interpretação deve, portanto, ser guiada por conhecimento físico e metalúrgico, avaliando se os padrões aprendidos são plausíveis à luz de mecanismos conhecidos. Em imagens, essa preocupação é ainda mais relevante, pois redes convolucionais podem aprender artefatos de aquisição, padrões de iluminação ou ruídos sistemáticos se o conjunto de dados não for representativo do ambiente real. A generalização para novas condições experimentais só é possível quando o conjunto de treino captura adequadamente a variabilidade esperada em uso.

Por fim, o aprendizado de máquina em ciência dos materiais deve ser entendido como ferramenta integrada ao método científico, e não como substituto da modelagem física ou da experimentação. Seu maior valor está em organizar grandes volumes de dados, identificar tendências, agrupar comportamentos semelhantes, priorizar experimentos e automatizar tarefas repetitivas de inspeção e análise. Modelos bem utilizados aceleram o ciclo de desenvolvimento de materiais, reduzem o espaço de busca experimental e auxiliam na tomada de decisão, desde que seus limites sejam claramente reconhecidos. A validação física, a revisão crítica dos resultados e a coerência com princípios fundamentais permanecem como critérios finais de aceitação, garantindo que o aprendizado de máquina contribua de forma robusta, responsável e cientificamente sólida para a área de ciência dos materiais.
