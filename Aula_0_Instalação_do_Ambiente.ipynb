
# Tutorial de Instalação do Python para Windows e macOS

Este notebook apresenta instruções completas para instalar Python 3.10 e a versão mais recente do Python em **Windows** e **macOS**, incluindo configuração de PATH, instalação do pip e verificação do funcionamento. Não envolve configuração avançada no VS Code, pois assume que a extensão Python já resolve o ambiente.

---

## 1. Instalação do Python no Windows

1. Acesse *https://www.python.org/downloads/* e baixe a versão mais recente.
2. Baixe também a versão **3.10** em *https://www.python.org/downloads/release/python-3100/*.
3. Execute cada instalador separadamente.
4. Na primeira tela, marque **Add Python to PATH**.
5. Clique em **Install Now**.

### Verificação no Prompt

Abra o Prompt de Comando e execute:

```
python --version
pip --version
py -3.10 --version
py -3.12 --version
```

Se o pip não funcionar, configure o PATH manualmente:

Adicione no ambiente:
```
C:\Users\SEU_USUARIO\AppData\Local\Programs\Python\Python3XX\
C:\Users\SEU_USUARIO\AppData\Local\Programs\Python\Python3XX\Scripts\
```

---

## 2. Instalação do Python no macOS

1. Instale o Homebrew:
```
/bin/bash -c "$(curl -fsSL https://raw.githubusercontent.com/Homebrew/install/HEAD/install.sh)"
```

2. Instale Python:
```
brew install python
```

3. Instale Python 3.10 (opções):
```
brew install pyenv
pyenv install 3.10.13
pyenv install 3.12.1
```

4. Verifique:
```
python3 --version
pip3 --version
```

---

# Instalação do Visual Studio Code no Windows e no macOS

## 1. Download e instalação do VS Code

O Visual Studio Code está disponível para Windows e macOS no endereço:

```
https://code.visualstudio.com/
```

Após acessar o site, selecione a versão correspondente ao sistema operacional e realize o download.

### Windows

1. Execute o instalador baixado.
2. Selecione a opção para adicionar o VS Code ao Path, se disponível.
3. Finalize a instalação.

### macOS

1. Abra o arquivo .zip baixado.
2. Arraste o aplicativo Visual Studio Code para a pasta Applications.
3. Abra o VS Code pela primeira vez para registrar permissões do sistema.

---

## 2. Instalação da extensão Python

Após instalar o VS Code:

1. Abra o VS Code.
2. Acesse o menu de extensões (ícone de quadrados à esquerda).
3. Busque por **Python** (Microsoft).
4. Instale a extensão.
5. O VS Code passará a reconhecer automaticamente instalações do Python, seja a versão mais recente ou a versão 3.10.

A extensão oferece suporte para execução de scripts, depuração, autocompletar, análise de código e integração com ambientes virtuais.

---

## 3. Seleção do interpretador Python

Quando um arquivo `.py` é aberto pela primeira vez:

1. O VS Code exibe uma notificação solicitando a escolha do interpretador.
2. Caso a notificação não apareça, pressione **Ctrl + Shift + P** e pesquise **Python: Select Interpreter**.
3. Selecione o Python desejado (versão mais recente ou 3.10).
4. O interpretador escolhido permanece associado ao workspace atual.

---

## 4. Verificação do funcionamento

Para garantir que o Python está configurado corretamente no VS Code:

1. Crie um arquivo `teste.py`.
2. Insira o conteúdo:

```python
print("Ambiente Python ativo no VS Code")
```

3. Execute o script pressionando **Run** no canto superior direito.
4. O terminal integrado deve exibir a mensagem.

Após esta configuração, o VS Code estará pronto para funcionar com Python, permitindo alternar entre diferentes versões conforme necessário para projetos de ciência dos materiais.
